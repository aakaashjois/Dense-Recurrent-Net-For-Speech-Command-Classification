```python
import numpy as np
import librosa
import librosa.display
import os
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
train_data = np.load(os.path.join('data', 'train_data_100.npy'))
```

```python
spec = librosa.feature.melspectrogram(train_data[0], sr=16000, fmax=3000)
spec_db = librosa.power_to_db(spec, ref=np.max)
```

```python
fig = plt.figure(figsize=(20, 10))
fig = librosa.display.specshow(spec_db, sr=16000, x_axis='time', y_axis='mel', fmax=3000)
plt.colorbar(format='%+02.0f dB')
plt.show()
```
